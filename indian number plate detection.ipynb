```python
test_samples = np.random.randint(0,len(npd),5)
drawn_samples = npd.drop(test_samples,axis = 0)
```

```python
WIDTH = 224
HEIGHT = 224
channel = 3

def show_img(index):
    image= imread("Indian Number Plate/" + npd["image_name"].iloc[index])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, dsize=(WIDTH, HEIGHT))
    
    tx = int(npd["top_x"].iloc[index] * WIDTH)
    ty = int(npd["top_y"].iloc[index] * HEIGHT)
    bx = int(npd["bottom_x"].iloc[index] * WIDTH)
    by = int(npd["bottom_y"].iloc[index] * HEIGHT)

    image = cv2.rectangle(image, (tx, ty), (bx, by), (0, 0, 255), 1)
    plt.imshow(image)
    plt.show()
```

```python
# from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_dataframe(
    drawn_samples,
    directory="Indian Number Plate/",
    x_col="image_name",
    y_col=["top_x", "top_y", "bottom_x", "bottom_y"],
    target_size=(WIDTH, HEIGHT),
    batch_size=32, 
    class_mode="other",
    subset="training")

validation_generator = datagen.flow_from_dataframe(
    drawn_samples,
    directory="Indian Number Plate/",
    x_col="image_name",
    y_col=["top_x", "top_y", "bottom_x", "bottom_y"],
    target_size=(WIDTH, HEIGHT),
    batch_size=32, 
    class_mode="other",
    subset="validation")
```

```python
from keras.applications.vgg16 import VGG16
```

```python
from keras.models import Model, Sequential
```

```python
from keras.layers import Input, Dense
```

```python
from keras.layers import Flatten, Dense, Conv2D, MaxPooling2D, Input, Dropout
```

```python
model = Sequential()
model.add(VGG16(weights="imagenet", include_top=False, input_shape=(HEIGHT, WIDTH, channel)))
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(4, activation="sigmoid"))

model.layers[-6].trainable = False

model.summary()
```

```python
STEP_SIZE_TRAIN = int(np.ceil(train_generator.n / train_generator.batch_size))
STEP_SIZE_VAL = int(np.ceil(validation_generator.n / validation_generator.batch_size))

print("Train step size:", STEP_SIZE_TRAIN)
print("Validation step size:", STEP_SIZE_VAL)

train_generator.reset()
validation_generator.reset()
```

```python
from keras.optimizers import sgd
```

```python
sgd = sgd(lr=0.0005)
model.compile(optimizer=sgd, loss="mse")
```

```python
train_modelling = model.fit_generator(train_generator,
    steps_per_epoch=STEP_SIZE_TRAIN,
    validation_data=validation_generator,
    validation_steps=STEP_SIZE_VAL,
    epochs=30)
```

## test

```python
model.evaluate_generator(validation_generator, steps=STEP_SIZE_VAL)
```

```python
for idx, row in npd.iloc[test_samples].iterrows():    
    img = cv2.resize(cv2.imread("Indian Number Plate/" + row[0]) / 255.0, dsize=(WIDTH, HEIGHT))
    y_hat = model.predict(img.reshape(1, WIDTH, HEIGHT, 3)).reshape(-1) * WIDTH
    
    xt, yt = y_hat[0], y_hat[1]
    xb, yb = y_hat[2], y_hat[3]
    
    img = cv2.cvtColor(img.astype(np.float32), cv2.COLOR_BGR2RGB)
    image = cv2.rectangle(img, (xt, yt), (xb, yb), (0, 0, 255), 1)
    plt.imshow(image)
    plt.show()
```
