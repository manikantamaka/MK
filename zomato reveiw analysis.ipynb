## converting rating into float it is in string

```python
a = zomato
a.rate = a.rate.astype(str)
a.rate = a.rate.apply(lambda a :a.replace('/5',''))
a.rate = a.rate.apply(lambda a:float(a))
```

## stacked Bar  

### used for show comparisons between categories of data

```python
plt.rcParams['figure.figsize'] = (15, 9)
x = pd.crosstab(a['rate'], a['online_order'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True,color=['black','orange'])
plt.title('online order vs rate')
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(15,8))
x = pd.crosstab(a['rate'], a['book_table'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True,color=['black','yellow'])
plt.title('online order vs rate')
plt.legend(loc="upper right")
plt.show()
```
